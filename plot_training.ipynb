```python
import json
import os
import matplotlib.pyplot as plt
output_path = "outputs/lmrec_toronto_reproduce"
```

```python
with open(os.path.join(output_path, "trainer_state.json"), 'r') as fin:
    log_history = json.load(fin)['log_history']
```

```python
epoch = [log["epoch"] for log in log_history if "epoch" in log and "eval_loss" not in log and "train_runtime" not in log]
loss = [log["loss"] for log in log_history if "loss" in log]
eval_epoch = [log["epoch"] for log in log_history if "epoch" in log and "eval_loss" in log]
eval_loss = [log["eval_loss"] for log in log_history if "eval_loss" in log]
eval_accuracy = [log["eval_accuracy"] for log in log_history if "eval_accuracy" in log]
```

```python
plt.plot(epoch, loss)
plt.plot(eval_epoch, eval_loss)
```

```python
plt.plot(eval_epoch, eval_accuracy)
```
